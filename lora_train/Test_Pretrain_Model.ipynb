```python
import torch
from diffusers import DiffusionPipeline
import os
```

```python
# Percorso del modello che hai appena scaricato
model_path = os.path.expanduser("~/models/sdxl-base")
output_path = os.path.expanduser("~/prog/output_PTM/Random_img/")
filename = "astronaut_riding_cat.png"
output_file = os.path.join(output_path, filename)

print("🚀 Caricamento SDXL in corso...")
pipe = DiffusionPipeline.from_pretrained(
    model_path, 
    torch_dtype=torch.float16, 
    variant="fp16", 
    use_safetensors=True
)

# 1. Sequential CPU Offload: È più aggressivo del normale offload.
# Carica sulla GPU UN SOLO LAYER alla volta. È lento, ma permette di 
# far girare SDXL anche dove normalmente non starebbe.
pipe.enable_sequential_cpu_offload()

# 2. VAE Tiling: Indispensabile per non crashare alla fine.
pipe.vae.enable_tiling()

# 3. Xformers: Fondamentale per le schede serie 2000.
pipe.enable_xformers_memory_efficient_attention()


prompt = "A realistic photo of an Astronaut riding a cat"

print("🎨 Generazione immagine...")
# Usiamo 25 step invece di 30 per velocizzare il test
image = pipe(prompt=prompt).images[0]

# Salviamo con un nome diverso per non fare confusione
image.save(output_file)
print("✅ Immagine salvata correttamente!")
```
