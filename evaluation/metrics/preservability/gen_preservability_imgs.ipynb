```python
import torch
from diffusers import DiffusionPipeline, StableDiffusionXLPipeline
import os
import gc
import yaml

with open(os.path.expanduser("~/prog/config.yaml"), "r") as f:
    config = yaml.safe_load(f)
```

```python
MODEL_PATH = config["paths"]["base_model_dir"]
LORA_PATH = config["paths"]["lora_model_dir"]
LORA_WEIGHTS_FILE = config["weights_file"]["lora_weights"]
OUTPUT_DIR_LORA = config["paths"]["evaluation_dir"] + "/metrics/preservability/images_lora"
OUTPUT_DIR_WLORA = config["paths"]["evaluation_dir"] + "/metrics/preservability/images_base"

OUTPUT_LORA_CAT = os.path.join(OUTPUT_DIR_LORA, "cats")
OUTPUT_WLORA_CAT = os.path.join(OUTPUT_DIR_WLORA, "cats")

NUM_IMAGES = 5 # Number of images per prompt

INFERENCE_STEPS = 50

prompts = [
    "A photo of a cat",
    "A photo of a dog",
    "A photo of a rabbit",
    "A photo of a fox",
    "A photo of a lion",
    "A photo of a tiger",
    "A mountain landscape",
    "A portrait of a person",
    "A city skyline",
    "A bowl of fruit",
    "A car on a road"
]

seeds = [65, 156, 2004, 11234, 202021] # Fixed seeds for reproducibility
```

## Load Model LoRA

```python
print("Loading Lora model...")

pipeLora = DiffusionPipeline.from_pretrained(
    MODEL_PATH, 
    torch_dtype=torch.float16, 
    variant="fp16", 
    use_safetensors=True
).to("cuda")  # Carica il modello base in FP16 per risparmiare memoria GPU

pipeLora.load_lora_weights(
    LORA_PATH,
    weight_name=LORA_WEIGHTS_FILE
)

for i, prompt in enumerate(prompts):
    for j in range(NUM_IMAGES):
        image_num = j+1
        current_seed = seeds[j]
        generator = torch.Generator("cuda").manual_seed(current_seed)

        filename = f"image_{i+1:02d}_{current_seed}.png"
        filepath = os.path.join(OUTPUT_DIR_LORA, filename)

        print(f"Generating image for prompt '{prompt}' with seed {current_seed}...")

        image = pipeLora(prompt, num_inference_steps=INFERENCE_STEPS, generator=generator).images[0]
        image.save(filepath)
```

## Clean cache

```python
gc.collect() # Clear memory before loading the base model without Lora
torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()
```

## Load Base Model

```python
print("Loading base model...")

pipe = DiffusionPipeline.from_pretrained(
    MODEL_PATH, 
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True
).to("cuda")  

for i, prompt in enumerate(prompts):
    for j in range(NUM_IMAGES):
        image_num = j+1
        current_seed = seeds[j]
        generator = torch.Generator("cuda").manual_seed(current_seed)

        filename = f"image_{i+1:02d}_{current_seed}.png"
        filepath = os.path.join(OUTPUT_DIR_WLORA, filename)

        print(f"Generating image for prompt '{prompt}' with seed {current_seed}...")

        image = pipe(prompt, num_inference_steps=INFERENCE_STEPS, generator=generator).images[0]
        image.save(filepath)
```

## Generate images for Preservability
Generate 50 images of a random cat with Base Model and LoRA Model, starting with the same seeds (1,...,50)

```python
p = prompts[0]
num_images = 50
seeds = range(1,num_images+1)

def generate_set(pipe, output_path):
    for seed in seeds:
        generator = torch.Generator("cuda").manual_seed(seed)
        image = pipe(p, num_inference_steps=INFERENCE_STEPS, generator=generator).images[0]
        filename = f"cat_{seed}.png"
        filepath = os.path.join(output_path, filename)
        image.save(filepath)


pipe = DiffusionPipeline.from_pretrained(
    MODEL_PATH, 
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True
).to("cuda")  

#generate_set(pipe, OUTPUT_WLORA_CAT)

del pipe
gc.collect()
torch.cuda.empty_cache()

pipeLora = DiffusionPipeline.from_pretrained(
    MODEL_PATH, 
    torch_dtype=torch.float16, 
    variant="fp16", 
    use_safetensors=True
).to("cuda") 

pipeLora.load_lora_weights(
    LORA_PATH,
    weight_name=LORA_WEIGHTS_FILE
)
generate_set(pipeLora, OUTPUT_LORA_CAT)
```
